```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_path = '/content/drive/MyDrive/Prodigy_Internship_DataScience/weather_data.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to understand the structure
print(df.head())
```

```python
plt.figure(figsize=(5, 3))
plt.hist(df['Temperature_C'], bins=6, edgecolor='black')
plt.title('Distribution of Temperatures in San Diego')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.hist(df['Wind_Speed_kmh'], bins=5, edgecolor='black')
plt.title('Distribution of Wind Speeds in San Diego')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.hist(df['Date_Time'], bins=5, edgecolor='black')
plt.title('Distribution of Temperature in San Diego by date')
plt.xlabel('date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```
